```python
from ultralytics import YOLO

# 1. 훈련으로 생성된 가장 좋은 모델('best.pt') 로드
# OBB 모델은 'runs/obb/train/weights/best.pt' 경로에 저장됩니다.
# 만약 훈련 세션을 여러 번 실행했다면 'train2', 'train3' 등으로 폴더명이 바뀔 수 있습니다.
model = YOLO('runs/obb/train/weights/best.pt')

# 2. 모델 성능 검증
# 훈련 시 사용했던 dataset의 위치를 참조하여 data.yaml 파일을 지정합니다.
# model.val()은 검증 데이터셋(valid set)으로 모델을 평가합니다.
metrics = model.val(data='acne-3/data.yaml')

# 3. 주요 성능 지표 출력
print("성능 지표 (metrics):")
print(f"mAP50-95(B): {metrics.box.map}")   # mAP at IoU 0.50-0.95
print(f"mAP50(B): {metrics.box.map50}") # mAP at IoU 0.50
print(f"Precision(B): {metrics.box.p[0]}") # 첫 번째 클래스의 Precision
print(f"Recall(B): {metrics.box.r[0]}")    # 첫 번째 클래스의 Recall
```
